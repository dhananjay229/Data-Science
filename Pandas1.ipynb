```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]})
df['States'] = "CA NV AZ".split()
df.set_index('States',inplace = True)
print(df)
```

```python
print(df.dropna(axis = 1))
```

```python
print(df.fillna(value="FILL NA"))
```

```python
print(df.fillna(value=df["A"].mean()))
```

```python
data = {'Company':['GOOG','GOOG','MFST','MFST','FB','FB','Apple'],
     'Person':['sam','charly','dk','sh','ss','mg','ak'],
     'Sales':[200,300,400,500,600,700,800]}
df = pd.DataFrame(data)
print(df)
```

```python
group = df.groupby('Company')
print(group.mean())
```

```python
print(group.sum())
```

```python
pd.DataFrame(df.groupby('Company').describe().loc['FB']).transpose()
```

```python
pd.DataFrame(df.groupby('Company').describe().loc[['FB','GOOG']])
```

```python
left = pd.DataFrame({'key':['K0','K8','K2','K3'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']
                    })
right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                     'C':['C0','C1','C2','C3'],
                     'D':['D0','D1','D2','D3']
                     })
```

```python
left
```

```python
right
```

```python
merge1 = pd.merge(left,right,how = 'inner',on = 'key')
merge1
```

```python
left = pd.DataFrame({'key1':['K0','K0','K1','K2'],
                     'key2':['K0','K1','K0','K1'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']
                    })
right = pd.DataFrame({'key1':['K0','K1','K1','K2'],
                      'key2':['K0','K0','K0','K0'],
                     'C':['C0','C1','C2','C3'],
                     'D':['D0','D1','D2','D3']
                     })
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on=['key1','key2'])
```

```python
pd.merge(left,right,how='left',on=['key1','key2'])
```

```python
pd.merge(left,right,how='right',on=['key1','key2'])
```

```python
left = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']},
                     index = ['K0','K1','K2','K3']
                    )

right = pd.DataFrame({'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']},
                     index = ['K0','K1','K2','K3']
                    )
```
