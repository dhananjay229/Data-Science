# Seaborn

1. Numerical Data Ploting

```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
sns.set(style = 'darkgrid')
```

```python
tips = pd.read_csv('F:\python/tips.csv')
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style="time")
```

```python
sns.relplot(x='total_bill',y='tip', hue='smoker',col='time',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip', hue='smoker',col='size',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip', hue='smoker',col='time',col_wrap=3,height=3,data=tips)
```

Categorical data ploting

```python
sns.countplot('sex',data=tips)
```

```python
sns.countplot(y='sex',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.boxplot(x='smoker',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=tips)
```

```python
tips.head()
```

```python
sns.violinplot(x='total_bill',y='day',palette='rainbow',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm', hue = 'size')
```

Distribution Plot

```python
sns.distplot(tips['tip'])
```

```python
sns.distplot(tips['tip'],kde =False ,bins=10)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(data=tips,hue='sex')
```

Linear Regression and Relationship

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x = 'total_bill', y= 'tip', data = tips)
```
